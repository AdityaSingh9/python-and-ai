## text -> vectors using bow(one of the average approach becoz its good for small datasets and task)

```python
import pandas as pd
messages=pd.read_csv('/Users/adityakumarsingh/python/smsspam.csv')
```

```python
print(messages)
```

```python
#data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
## This data cleaning and pre processing invloves removing spcl char, lower casing words , applying stop words and creating a corpus/paragraph
## this is a generic approach and can be applied to any data set
corpus=[]
for i in range(0,len(messages)):
    text = messages["'text'"][i]
    if pd.isna(text):  # skip if text is NaN
        continue
    review = re.sub('[^a-zA-Z]', ' ', text)  # fix typo: A-z → A-Z
    review=review.lower() #lower casing the strings
    review=review.split() #converting a string into list of words to apply stop words
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]  #applying stop words
    review=" ".join(review)
    corpus.append(review)
```

```python
print(corpus)
```

```python
##create BOW:
# 'type' is output variable and 'text' is input feature:
from sklearn.feature_extraction.text import CountVectorizer #it has attribute present for stop words and lowercasing
cv=CountVectorizer(max_features=100,binary=True)   #take these many words wihch have max freq
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
##create BOW with N-GRAMS
# 'type' is output variable and 'text' is input feature:
from sklearn.feature_extraction.text import CountVectorizer #it has attribute present for stop words and lowercasing
cv=CountVectorizer(max_features=105,binary=True,ngram_range=(2,2))   #use any ngram to increase vocabulary 
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import pandas as pd
messages=pd.read_csv('/Users/adityakumarsingh/python/smsspam.csv')
#data cleaning and preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordLlemmatize=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    text = messages["'text'"][i]
    if pd.isna(text):  # skip if text is NaN
        continue
    review = re.sub('[^a-zA-Z]', ' ', text)  # fix typo: A-z → A-Z
    review=review.lower() #lower casing the strings
    review=review.split() #converting a string into list of words to apply stop words
    review=[wordLlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]  #applying stop words
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
##TF-IDF and N-GRAMS
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x: "%.3g" % x)) #just to print formatted array
```
